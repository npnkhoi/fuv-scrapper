```python
import pandas as pd
```

```python
data = pd.read_csv('assets/sport.csv')
```

```python
data
```

```python
js = []

for i in range(len(data)):
    row = data.iloc[i, :]
    js.append({
        'id': 'CLB' + str(i).zfill(3),
        'title': row['Club'],
        'location': row['Location'],
        'schedule': [] if not isinstance(row['Schedule'], str) else [
            {
                'day': s.split(',')[0],
                'start_time': s.split(' ')[1] + ':00',
                'end_time': s.split(' ')[3] + ':00',
            }
            for s in row['Schedule'].split('\n')
        ],
        "credits": "0",
        "instructor": "Coaches",
        "url": "",
        "description": "N/A",
        "categories": [],
    })
js
```

```python
import json
json.dump(js, open('assets/sport.json', 'w+'))
```
